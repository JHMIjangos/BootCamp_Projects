```python
import pandas as pd
import matplotlib.pyplot as plt
import os, sys
import numpy as np
import warnings
import seaborn as sns
from scipy import signal
from scipy import fftpack
from scipy import signal
import statistics


# Ajustar el nombre de las columnas en el Notebook de ser necesario
path= r'202208291716_pmxnt10002mx.csv'
cols = ['mem_util', 'timestamp', 'cpu_util']                                            
freq_data = '5min'


class Warning:
    def __init__(self,path,cols,freq_data,serie):
       df = pd.read_csv(path)
       df = df[cols].drop_duplicates()
       df['timestamp'] = pd.to_datetime(df['timestamp'], dayfirst=True)
       df.set_index('timestamp', inplace = True)
       self.df = df.asfreq(freq_data)
       self.serie = serie
    
    def plot_serie(self):
        header = self.serie
        self.df[header].plot(figsize = (9,3))
        plt.title(header)
        plt.show()
        
    def completa_datos(self):
        check = self.df[self.serie].isnull().values.any()
        if check:
            count_nan=self.df[self.serie].isnull().sum()
            porc_obsint = count_nan/len(self.df[self.serie])
            self.df[self.serie] = self.df[self.serie].interpolate(method = 'linear')
            header = self.serie
            self.df[self.serie].plot(figsize = (9,3))
            plt.title(header)
            plt.show()
            print ("Se interpolaron un {0:.2f}% de las observaciones totales".format(porc_obsint*100))
        
        else:
            print(f"La serie no contiene valores nulos")
            header = self.serie
            self.df[self.serie].plot(figsize = (9,3))
            plt.title(header)
            plt.show()
            
    def t_fourier(self,tiempo,frec):
      #serie
      #
        y = self.df[self.serie].values.tolist()
        N = len(y)
        Ts = tiempo 
        t = np.linspace(0, (N-1)*Ts, N)
        Y = fftpack.fft(y)             #Tranformada de fourier
        Fs = frec                       
        f = fftpack.fftfreq(N)*Fs
        plt.figure(figsize = (9,3))
        plt.plot(f, np.abs(Y))
        plt.title(f"Transformacion de fourier a la serie " + self.serie)
        plt.show()

    
    def det_warning(self,escala, t_signal, Fs, n_desv, window, lim_m, umbral):
        b, a = signal.butter(1, escala, btype=t_signal, fs = Fs)
        y = self.df[self.serie].values.tolist()
        filtedData = signal.filtfilt(b,a,y)
        smothedData = pd.DataFrame({'timestamp': self.df.index,'y':y, 'smothed_y':filtedData})
        smothedData['smothed_y_upper'] = smothedData['smothed_y']+statistics.stdev(smothedData['smothed_y']*n_desv)
        smothedData['m'] = np.nan
        for k in smothedData.index[:-1]:
            smothedData['m'][k+1] = smothedData['smothed_y'][k+1]-smothedData['smothed_y'][k]
        smothedData.head()
        wind = []
        winds_sum = []
        for k in range(window):
            wind.append(np.nan)
        for k in smothedData.index[window:]:
            wind.append([smothedData['m'][k-window+1:k+1].values])
        smothedData['wind'] = wind
        for k in smothedData.wind:
            try:
                winds_sum.append(sum(k[0]))
            except:
                winds_sum.append(0)
        smothedData['winds_sum'] = winds_sum
        smothedData['warning'] = np.nan
        for k in smothedData.index:
            if (smothedData['winds_sum'][k]>lim_m) or (smothedData['smothed_y_upper'][k]>umbral):
          #   if smothedData['winds_sum'][k]>p:
                smothedData['warning'][k] = 1
            else:
                smothedData['warning'][k] = 0
        smothedData['y_warnings'] = df[(df['warning']==1)]
        smothedData['smothed_y_upper'].plot(figsize = (9,3))
        plt.show()
        plt.title(self.serie)
        smothedData.to_csv("warning_signals.csv") 


memoria = Warning(path, cols,freq_data,'mem_util')
memoria.plot_serie()
memoria.completa_datos()
memoria.t_fourier(1/300,300)
memoria.det_warning(2,'lowpass',300,1,3,0.3,80)
```
